```python
# Importing necessary libraries: TensorFlow and TensorFlow Datasets
import tensorflow as tf
import tensorflow_datasets as tfds

# Loading the MNIST dataset, a collection of handwritten digits
# The dataset is split into training and test sets
# shuffle_files=True ensures the files are shuffled to prevent sequence bias
# as_supervised=True loads the dataset in a supervised format (input, label)
# with_info=True provides a metadata object with info about the dataset
(ds_train, ds_test), ds_info = tfds.load(
    'mnist',
    split=['train', 'test'],
    shuffle_files=True,
    as_supervised=True,
    with_info=True,
)

# Define a normalization function for the images
# This function converts image data from uint8 to float32 and normalizes it
def normalize_img(image, label):
  """Normalizes images: `uint8` -> `float32`."""
  return tf.cast(image, tf.float32) / 255., label

# Apply the normalization function to the training dataset
# num_parallel_calls=tf.data.AUTOTUNE optimizes the number of parallel calls
ds_train = ds_train.map(
    normalize_img, num_parallel_calls=tf.data.AUTOTUNE)

# Cache the dataset to improve performance
ds_train = ds_train.cache()

# Shuffle the training dataset using the number of examples as the shuffle buffer size
ds_train = ds_train.shuffle(ds_info.splits['train'].num_examples)

# Batch the dataset into batches of 128 images
ds_train = ds_train.batch(128)

# Prefetch the data for optimal GPU utilization
ds_train = ds_train.prefetch(tf.data.AUTOTUNE)

# Apply the normalization and batching to the test dataset as well
ds_test = ds_test.map(
    normalize_img, num_parallel_calls=tf.data.AUTOTUNE)
ds_test = ds_test.batch(128)
ds_test = ds_test.cache()
ds_test = ds_test.prefetch(tf.data.AUTOTUNE)

# Define a Sequential model using Keras
# The model consists of a Flatten layer, a Dense layer with ReLU activation, and an output Dense layer
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),  # Flattening the 28x28 images
  tf.keras.layers.Dense(128, activation='relu'),  # Dense layer with 128 neurons and ReLU activation
  tf.keras.layers.Dense(10)                       # Output layer with 10 neurons (one for each digit)
])

# Compile the model with the Adam optimizer, Sparse Categorical Crossentropy loss, and accuracy as a metric
model.compile(
    optimizer=tf.keras.optimizers.Adam(0.001),   # Using Adam optimizer with a learning rate of 0.001
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),  # Loss function for multi-class classification
    metrics=[tf.keras.metrics.SparseCategoricalAccuracy()],  # Metric to monitor
)

# Train the model on the training dataset
# Validate it against the test dataset
model.fit(
    ds_train,
    epochs=6,  # Training for 6 epochs
    validation_data=ds_test,  # Using the test dataset for validation
)
```

```python
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
import numpy as np

# Load the MNIST dataset
(ds_train, ds_test), ds_info = tfds.load(
    'mnist',
    split=['train', 'test'],
    shuffle_files=True,
    as_supervised=True,
    with_info=True,
)

# Define the normalization function
def normalize_img(image, label):
  """Normalizes images: `uint8` -> `float32`."""
  return tf.cast(image, tf.float32) / 255., label

# Applying the normalization function to the datasets
ds_train = ds_train.map(normalize_img, num_parallel_calls=tf.data.AUTOTUNE).cache().shuffle(ds_info.splits['train'].num_examples).batch(128).prefetch(tf.data.AUTOTUNE)
ds_test = ds_test.map(normalize_img, num_parallel_calls=tf.data.AUTOTUNE).batch(128).cache().prefetch(tf.data.AUTOTUNE)

# Define a Sequential model with a dropout layer
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(64, activation='relu'),
  #tf.keras.layers.Dropout(0.1),  # Dropout layer to prevent overfitting
  tf.keras.layers.Dense(10, activation='relu')
])

# Compile the model
model.compile(
    optimizer=tf.keras.optimizers.Adam(0.001),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=[tf.keras.metrics.SparseCategoricalAccuracy()],
)

# Train the model
model.fit(
    ds_train,
    epochs=10,
    validation_data=ds_test,
)

# Function to display wrongly predicted images
def display_wrong_predictions(test_dataset, num_samples=3):
    wrong_samples = 0
    plt.figure(figsize=(10, 4))

    for images, labels in test_dataset.unbatch().batch(1):
        prediction = model.predict(images)
        predicted_label = np.argmax(prediction, axis=1)[0]

        if predicted_label != labels.numpy():
            wrong_samples += 1

            # Display original image with original label
            plt.subplot(2, num_samples, wrong_samples)
            plt.imshow(tf.squeeze(images))
            plt.title(f"Orig: {labels.numpy()}")
            plt.axis("off")

            # Display original image with predicted label
            plt.subplot(2, num_samples, num_samples + wrong_samples)
            plt.imshow(tf.squeeze(images))
            plt.title(f"Pred: {predicted_label}")
            plt.axis("off")

            if wrong_samples == num_samples:
                break

    plt.show()

# Display 5 wrongly predicted images
print("Displaying wrongly predicted images:")
display_wrong_predictions(ds_test)
```
